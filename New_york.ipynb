```python
data['id']=data['id'].astype(object)

data['host_id']=data['host_id'].astype(object)
```

```python
data['price']
```

```python
df=data[data["price"]<1500]

plt.figure(figsize=(4,4))
sns.boxplot(x="price",data=df)
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.histplot(x="price",data=df,bins=40)
plt.ylabel("Frequency")
plt.title("Price Distribution")
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.histplot(x="availability_365",data=df,bins=30)
plt.ylabel("Frequency")
plt.title("Availability_365 Distribution")
plt.show()
```

```python
df.columns
```

```python
df.groupby(by="neighbourhood_group")['price'].mean()
```

```python
df.head()
```

```python
df['price per bed'] = df['price'] / df['beds'] 
df.head()  
```

```python
df.groupby(by="neighbourhood_group")['price per bed'].mean()
```

```python
df.columns
```

```python
plt.figure(figsize=(4,4))
sns.barplot(x="neighbourhood_group",y="price",data=df,hue="room_type")
plt.xticks(rotation=75)
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.scatterplot(x="number_of_reviews",y="price",data=df,hue="neighbourhood_group")
plt.title("Locality and Review Dependency")
plt.show()
```

```python
df.info()
```

```python
sns.pairplot(data=df,vars=["price","number_of_reviews","availability_365","minimum_nights"],hue="room_type")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="longitude",y="latitude",data=df,hue="room_type")
plt.title("Locality and Review Dependency")
plt.show()
```

```python
df.dtypes
```

```python
corr=df[["latitude","longitude","minimum_nights","number_of_reviews","reviews_per_month","availability_365","beds"]].corr()
corr
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(data=corr,annot=True)
plt.show()
```
